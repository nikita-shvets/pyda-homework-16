```python
import pandas as pd
import numpy as np
df = pd.read_csv('horse_data.csv',na_values = '?')
df.columns = ['surgery', 'Age', ' Hospital Number', 'rectal_temperature', 'pulse', 'respiratory_rate',
'temperature_of_extremities','peripheral_pulse', ' mucous membranes', 'capillary refill time', 'pain', 'peristalsis',
'abdominal distension', 'nasogastric tube', 'nasogastric reflux', ' nasogastric reflux PH', 'rectal examination', 'abdomen',
'packed cell volume', 'total protein', 'abdominocentesis appearance', 'abdomcentesis total protein', 'outcome',
'surgical lesion?', 'type of lesion_1', 'type of lesion_2', 'type of lesion_3', 'cp_data']
df = df.iloc[:,0:8]
df.describe()
#сразу видно что в некоторых столбцах есть пропуски, метрики имеют смысл для количественных данных(кроме процентилей)
# не в курсе че еще тут можно сказать), выбросы будем считать попозже
```

```python
#ищем выбросы, для тех столбцов где это имеет смысл(там где данные количественные)
#начнем с возраста
q1 = df.Age.quantile(0.25)
q3 = df.Age.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df.Age.between(lower_bound, upper_bound, inclusive=True)]
remove_outliers
pd.concat([df, remove_outliers]).drop_duplicates(keep=False)
#очевидно это не выброс, а недостаточно большая выборка данных, так как в выбросах всего одно значение возраста(9) и встречается
#достаточно часто 
```

```python
# так как в колонке surgery  у нас всего один пропуск убираем строку с пропуском(остальных данных будет достаточно)
df = df.drop(df[df.surgery.isnull()].index)
df
```

```python
# значения температур и пульса будем заменять медианной, учитвая была операция или нет
df.rectal_temperature.fillna(df.groupby('surgery').rectal_temperature.transform('median'), inplace=True)
df.pulse.fillna(df.groupby('surgery').pulse.transform('median'), inplace=True)
df.temperature_of_extremities.fillna(df.groupby('surgery').temperature_of_extremities.transform('median'), inplace=True)
df.peripheral_pulse.fillna(df.groupby('surgery').peripheral_pulse.transform('median'), inplace=True)
# а рейтинг репозитория просто заменим медианой
df.respiratory_rate.fillna(df.respiratory_rate	.median(), inplace=True)
df
```
